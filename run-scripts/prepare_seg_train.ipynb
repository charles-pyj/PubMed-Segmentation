```python
import os
import shutil

def copy_text_files(src_dir, dest_dir):
    # Create destination directory if it doesn't exist
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    # Walk through the source directory
    for root, dirs, files in os.walk(src_dir):
        for file in files:
            if file.endswith('.txt'):
                # Construct full file path
                file_path = os.path.join(root, file)
                # Copy file to destination directory
                shutil.copy2(file_path, dest_dir)
                print(f"Copied: {file_path} to {dest_dir}")

# Define source and destination directories
src_directory = './seg_resume'
dest_directory = './train_seg_all_text'

# Call the function to copy text files
copy_text_files(src_directory, dest_directory)
```
