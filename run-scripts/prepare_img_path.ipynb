```python
import json

# Path to the JSON file
file_path = '/nfs/turbo/umms-drjieliu/proj/medlineKG/data/PMC_figure/test_set/test.json'

# Open and read the JSON file
with open(file_path, 'r') as file:
    data = json.load(file)

print(len(data))
print(data[0])
```

```python
base_url = "https://openi.nlm.nih.gov"
img_paths = [base_url + i['imgLarge'] for i in data]
```

```python
import json

def chunk_list(data, chunk_size):
    """Split data into chunks of specified size."""
    for i in range(0, len(data), chunk_size):
        yield data[i:i + chunk_size]
```

```python
with open("./seg_train_img_url.json",'w') as f:
    json.dump(img_paths[:30000],f)
```

```python
import requests
from tqdm import tqdm
for i in tqdm(range(4253,30000)):
    path = img_paths[i]
    responseImg = requests.get(path)
    if responseImg.status_code == 200:
        #print("Success!")
        image_data = responseImg.content
        
        with open(f'./seg_resume/{i}.jpg', 'wb') as file:
            file.write(image_data)
            #print("Image saved successfully!")
    else:
        print("Failed to retrieve the image.")
```

```python
import os

def save_chunks_to_json(data, chunk_size, output_dir):
    """Split data into chunks and save each chunk to a separate JSON file."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        
    for i, chunk in enumerate(chunk_list(data, chunk_size)):
        file_path = os.path.join(output_dir, f'chunk_{i+1}.json')
        with open(file_path, 'w') as f:
            json.dump(chunk, f, indent=4)
        print(f"Chunk {i+1} saved to {file_path}")

# Example usage: # Example list with 200 image paths
chunk_size = 20000
output_dir = '/nfs/turbo/umms-drjieliu/proj/medlineKG/data/figure_json_by_article/chunks'

save_chunks_to_json(img_paths, chunk_size, output_dir)
```
